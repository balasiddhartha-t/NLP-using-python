```python
import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
articleURL = "https://www.washingtonpost.com/world/europe/companies-struggle-to-recover-after-massive-cyberattack-with-ransom-demands/2017/06/28/4e9e8fbe-5bd5-11e7-8e2f-ef443171f6bd_story.html?utm_term=.e648a0e5be2c"
```

```python
page = urlopen(articleURL).read().decode('utf8','ignore').replace('\n',' ').replace('\t','').replace('\s','')
soup = BeautifulSoup(page,'lxml')
```

###### Usually all the text in the Washington post is enclosed in the article tag. so we are returning the article

```python
soup.find('Article')
soup
```

###### Returns all the article elements

```python
text = ''. join(map(lambda p:p.text, soup.find_all('article')))
text
```

#### removes all the special Characters from the text

```python
import re
re.sub(r'[^\x00-\x7f]',r'',text)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
from nltk.corpus import stopwords
```

```python
from string import punctuation
```

```python
sents = sent_tokenize(text)
sents
```

```python
words_sent = word_tokenize(text.lower())
words_sent
```

```python
_stopwords = set(stopwords.words('english')+list(punctuation))
_stopwords
```

```python
words_sent = [word for word in words_sent if word not in _stopwords]
```

```python
words_sent
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(words_sent)
freq
```

```python
from heapq import nlargest
```

###### here in the below code 10 means the number of elements that are to be returned
###### freq is the collection that has to be sorted
###### the freq.get is the method to find the corresponding value in dict freq for a given key the function returns the count

```python
nlargest(10, freq, key = freq.get)
```

```python
from collections import defaultdict
```

```python
ranking= defaultdict(int)
for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
ranking
```

```python
sents_idx = nlargest(4, ranking , key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```
