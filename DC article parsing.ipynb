```python
import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
articleURL = "http://www.deccanchronicle.com/sports/cricket/110717/ravi-shastri-appointed-new-indian-cricket-coach-zaheer-khan-new-bowling-coach.html"
```

```python
page = urlopen(articleURL).read().decode('utf8','ignore').replace('\n',' ').replace('\t','').replace('\s','')
soup = BeautifulSoup(page,'lxml')
```

```python
soup
```

```python
soup.find('story-body')
soup
```

```python
text = ''. join(map(lambda p:p.text,soup.findAll("div", class_="story-body")))
text
```

```python
import re
text = re.sub(r'[^\x00-\x7f]',r'',text)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sents = sent_tokenize(text)
sents
```

```python
words_sent = word_tokenize(text.lower())
words_sent
```

```python
_stopwords = set(stopwords.words('english')+list(punctuation))
_stopwords
```

```python
words_sent = [word for word in words_sent if word not in _stopwords]
words_sent
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(words_sent)
freq
```

```python
from heapq import nlargest
```

```python
nlargest(10, freq, key = freq.get)
```

```python
from collections import defaultdict
```

```python
ranking= defaultdict(int)
for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
ranking
```

```python
sents_idx = nlargest(5, ranking , key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```
