```python
import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
articleURL = "http://www.thehindu.com/news/national/other-states/militants-open-fire-on-security-forces-in-anantnag/article19253160.ece?homepage=true"
```

```python
page = urlopen(articleURL).read().decode('utf8','ignore')
soup = BeautifulSoup(page,'lxml')
soup
```

```python
text = ''. join(map(lambda p:p.text, soup.find_all('p')))
text
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
from nltk.corpus import stopwords
from string import punctuation
```

```python
sents = sent_tokenize(text)
sents
```

```python
words_sent = word_tokenize(text.lower())
words_sent
```

```python
_stopwords = set(stopwords.words('english')+list(punctuation))
_stopwords
```

```python
words_sent = [word for word in words_sent if word not in _stopwords]
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(words_sent)
freq
```

```python
from heapq import nlargest
nlargest(10, freq, key = freq.get)
```

```python
from collections import defaultdict
```

```python
ranking= defaultdict(int)
for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
ranking
```

```python
sents_idx = nlargest(4, ranking , key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```
